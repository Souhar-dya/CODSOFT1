```python
from sarvamai import SarvamAI
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

client = SarvamAI(
    api_subscription_key="",
)

documents = []

# Document 1: Explain the economic impact of GST implementation in India.
print("Generating Document 1 (GST Impact)...")
response1 = client.chat.completions(
    model="sarvam-105b",
    messages=[
        {"role": "user", "content": "Explain the economic impact of GST implementation in India."}
    ],
    temperature=0.5,
    top_p=1,
    max_tokens=4000,
)
doc1_content = response1.choices[0].message.content
print("Document 1 (GST Impact) content excerpt:\n", doc1_content[:400], "...\n")
documents.append(doc1_content)

# Document 2: Describe the main challenges facing renewable energy adoption globally.
print("Generating Document 2 (Renewable Energy Challenges)...")
response2 = client.chat.completions(
    model="sarvam-105b",
    messages=[
        {"role": "user", "content": "Describe the main challenges facing renewable energy adoption globally."}
    ],
    temperature=0.5,
    top_p=1,
    max_tokens=4000,
)
doc2_content = response2.choices[0].message.content
print("Document 2 (Renewable Energy Challenges) content excerpt:\n", doc2_content[:400], "...\n")
documents.append(doc2_content)

# Document 3: Discuss the impact of artificial intelligence on the future of work.
print("Generating Document 3 (AI and Work)...")
response3 = client.chat.completions(
    model="sarvam-105b",
    messages=[
        {"role": "user", "content": "Discuss the impact of artificial intelligence on the future of work."}
    ],
    temperature=0.5,
    top_p=1,
    max_tokens=4000,
)
doc3_content = response3.choices[0].message.content
print("Document 3 (AI and Work) content excerpt:\n", doc3_content[:400], "...\n")
documents.append(doc3_content)

# Initialize TfidfVectorizer
# You can add parameters like stop_words='english' or adjust ngram_range for more specific analysis.
vectorizer = TfidfVectorizer(stop_words='english', max_features=100) # Limiting features for brevity

# Fit and transform the documents
tfidf_matrix = vectorizer.fit_transform(documents)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert TF-IDF matrix to a DataFrame for better readability
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
tfidf_df.index = [f"Document {i+1}" for i in range(len(documents))]

print("\nTF-IDF Matrix (Top 100 features, English stop words removed):")
display(tfidf_df)
```

### Step 1: Define a Search Query

First, let's define a query that we want to search within our generated documents.

```python
# Define the search query
search_query = "What are the main problems for solar and wind energy?"
print(f"Search Query: {search_query}")
```

### Step 2: Vectorize the Query and Calculate Similarity

Now, we'll vectorize the search query using the *same* `TfidfVectorizer` instance that was fitted on our documents. Then, we'll calculate the cosine similarity between the query vector and all document vectors to find the most relevant document.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Transform the query using the *fitted* vectorizer
query_vector = vectorizer.transform([search_query])

# Calculate cosine similarity between the query and all documents
similarity_scores = cosine_similarity(query_vector, tfidf_matrix)

# Get the index of the most similar document
most_similar_doc_index = similarity_scores.argmax()
most_similar_doc_score = similarity_scores[0, most_similar_doc_index]

print(f"Most similar document index: {most_similar_doc_index} (Document {most_similar_doc_index + 1})")
print(f"Similarity score: {most_similar_doc_score:.4f}")

# Retrieve the content of the most similar document
relevant_document_content = documents[most_similar_doc_index]

print(f"\nContent of the most relevant document (Document {most_similar_doc_index + 1} excerpt):\n{relevant_document_content[:500]}...")
```

### Step 3: Generate a Gist of the Relevant Document using SarvamAI

Finally, we'll take the content of the most relevant document and feed it to SarvamAI to generate a concise gist or summary.

```python
# Use SarvamAI to create a gist of the relevant document
print("Generating gist using SarvamAI...")

gist_prompt = f"Create a concise gist/summary of the following text:\n\n{relevant_document_content}"

gist = None # Initialize gist to None

try:
    gist_response = client.chat.completions(
        model="sarvam-105b",
        messages=[
            {"role": "user", "content": gist_prompt}
        ],
        temperature=0.3, # Lower temperature for more factual, less creative summary
        top_p=1,
        max_tokens=4000 # Increased max_tokens for desired gist length
    )

    # More robust check for content
    if gist_response and hasattr(gist_response, 'choices') and gist_response.choices and \
       hasattr(gist_response.choices[0], 'message') and gist_response.choices[0].message and \
       hasattr(gist_response.choices[0].message, 'content'):
        gist = gist_response.choices[0].message.content
        if gist is None or gist.strip() == "":
            print("\n--- SarvamAI API response had empty or None content ---")
            print(f"Full gist_response object: {gist_response}")
            gist = "Gist generation returned empty content."
    else:
        print("\n--- SarvamAI API response structure unexpected ---")
        print(f"Full gist_response object: {gist_response}")
        gist = "Gist generation failed: Unexpected API response structure."

except Exception as e:
    gist = f"An error occurred during gist generation: {e}"

print("\n--- Generated Gist ---")
print(gist)
print("\n--------------------")
```
